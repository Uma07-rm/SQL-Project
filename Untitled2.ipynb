```python
missing_values = das.isnull().sum()
missing_percentage = (missing_values / len(das)) * 100
```

```python
das_dropped = das.dropna()
```

```python
for col in das.columns:
    if das[col].dtype == 'object' or das[col].dtype == 'string':
        try:
            das[col] = pd.to_numeric(das[col], errors='unknown')
        except:
            pass
```

```python
numerical_cols = das.select_dtypes(include=[np.number]).columns

das[numerical_cols] = das[numerical_cols].apply(lambda x: x.fillna(x.mean()), axis=0)

print("\nDataFrame after Mean Imputation for Numerical Columns:")
print(das)
```

```python
non_numerical_cols = das.select_dtypes(exclude=[np.number]).columns

das[non_numerical_cols] = das[non_numerical_cols].apply(lambda x: x.fillna(x.mode()[0]), axis=0)

print("\nDataFrame after Handling Non-Numerical Columns:")
print(das)
```

```python
print("\nDataFrame after Handling All Columns:")
print(das)
```

```python
print("Original DataFrame:")
print(das)
```

```python
duplicates = das.duplicated()
print("\nDuplicate Rows:")
print(das[duplicates])
```

```python
das_no_duplicates = das.drop_duplicates()
```

```python
print("\nDataFrame after Removing Duplicates:")
print(das_no_duplicates)
```

```python
das.columns = das.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
print("Standardized DataFrame:")
print(das.head())
```

```python
print("Original DataFrame:")
print(das)

# Select only numerical columns
numerical_das = das.select_dtypes(include=[np.number])

# Z-scores for numerical columns
z_scores = np.abs(stats.zscore(numerical_das))
outliers = np.where(z_scores > 3)
print("\nOutliers detected using Z-score:")
print(numerical_das.iloc[outliers[0]])


das_no_outliers = numerical_das[(z_scores < 3).all(axis=1)]
print("\nDataFrame after removing outliers:")
print(das_no_outliers)

```

```python
das_capped = numerical_das.copy()
cap_threshold = 3
das_capped[(z_scores > cap_threshold)] = np.sign(das_capped) * cap_threshold * das_capped.std()

print("\nDataFrame after capping outliers:")
print(das_capped)
```

```python
das_log_transformed = numerical_das.applymap(lambda x: np.log(x) if x > 0 else x)
print("\nDataFrame after log transformation:")
print(das_log_transformed)


das_transformed_with_categorical = das_no_outliers.join(das.select_dtypes(exclude=[np.number]))

print("\nTransformed DataFrame with categorical columns re-integrated:")
print(das_transformed_with_categorical)
```
